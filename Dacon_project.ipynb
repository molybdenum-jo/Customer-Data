### favorita의 마케팅 성공사례

Favorita는 에콰도르의 대형 슈퍼마켓 중 하나로, 최근에는 다음과 같은 마케팅 전략을 사용해 성공적인 성과를 거두었습니다.

- 온라인 쇼핑 프로모션
Favorita는 인터넷과 모바일 앱을 통해 쇼핑몰을 운영하고 있습니다. 이러한 쇼핑몰에서는 주기적으로 할인 쿠폰이나 무료 배송 이벤트 등 다양한 프로모션을 진행합니다.

- 로열티 프로그램
Favorita는 로열티 프로그램인 "Mi Favorita"를 운영하고 있습니다. 이 프로그램은 매장에서 쇼핑을 할 때마다 포인트를 적립하고, 적립한 포인트를 현금처럼 사용할 수 있는 혜택을 제공합니다.

- 지역 커뮤니티 지원
Favorita는 지역 커뮤니티에서 다양한 활동을 지원하고 있습니다. 이를 통해 지역 사회와의 유대감을 형성하고, 고객들에게 좋은 인식을 전하는 마케팅 전략을 구사하고 있습니다.

- 친환경 제품 판매
Favorita는 친환경 제품 판매에도 힘을 쏟고 있습니다. 친환경 제품을 판매함으로써 소비자들에게 좋은 인식을 주고, 친환경적인 이미지를 강화하고 있습니다.

### favorita의 마케팅 실패사례
Favorita의 마케팅 실패 사례 중 하나는 2017 년에 발생한 것입니다. 그해 5 월 초, Favorita는 인터넷 쇼핑몰에서 상품을 주문하고 7 일 이내에 매장에서 수령 할 수있는 "Click & Collect" 서비스를 시작했습니다. 이 서비스는 대규모 프로모션 캠페인으로 지원되었으며, 고객들이 대량으로 방문하게되었습니다. 그러나 많은 고객들이 주문을하고 매장에 도착했을 때 상품이 준비되어 있지 않아 매우 실망했습니다. 이로 인해 Favorita는 신뢰도 문제와 함께 고객들의 실망을 초래하게 되어 이미지 손상이 있었습니다.

=> 고객들이 지역마다 매장내 상품물량을 확인할 수 있도록 해 해당 제품을 구매할 수 있도록 함(ex 올리브영 오늘드림 픽업서비스)
    
=> 지역별 잘 나가는 상품들을 예측하여 해당상품을 많이 가져다 놓음(ex 인구밀도별/ 문화적요소별)

### 강점:

높은 브랜드 인지도와 광고 효과: "Favorita"는 에콰도르에서 가장 잘 알려진 기업 중 하나로, 높은 브랜드 인지도와 광고 효과를 갖고 있습니다.
다양한 제품과 서비스: "Favorita"는 다양한 제품과 서비스를 제공하여 고객들에게 편리한 쇼핑 경험을 제공할 수 있습니다.
경쟁력있는 가격: "Favorita"는 경쟁 업체보다 더 경쟁력있는 가격을 제공할 수 있으며, 이는 고객들에게 매력적인 선택지가 됩니다.
다양한 위치와 출입문: "Favorita"는 도시와 시골 지역 모두에 많은 매장을 갖고 있으며, 이는 고객들이 쉽게 접근할 수 있도록 합니다.

### 약점:

경쟁 업체 증가: "Favorita"는 많은 경쟁 업체들과 경쟁하고 있으며, 경쟁이 더 심화될 가능성이 있습니다.
인프라 문제: 에콰도르의 일부 지역에서는 인프라가 발전하지 않아 "Favorita"가 확장하기 어려울 수 있습니다.
인력과 고객 서비스: "Favorita"의 일부 매장에서는 고객 서비스나 직원 교육 등에 대한 문제가 발생할 수 있습니다.
인터넷 기술 발전과의 경쟁: 온라인 쇼핑이 점차 증가하면서, "Favorita"는 이에 대한 대응과 경쟁에 대한 전략을 강화해야 할 필요가 있습니다.


- 상품군 변동성: "Favorita"는 다양한 상품군을 판매하고 있으며, 상품군별 판매량 변동성이 크다는 약점이 있을 수 있습니다. 이러한 약점은 재고 관리와 수익성 측면에서 부정적인 영향을 미칠 수 있습니다.
- 휴일/날씨 영향: 휴일이나 날씨 등의 외부 요인에 따라 "Favorita"의 매장들의 판매량이 감소할 수 있습니다. 이러한 상황에서는 적절한 예측과 대응 전략이 필요하며, 이를 위해 고급 분석 기술과 자원이 필요합니다.
- 기름값 상승: 에콰도르에서는 대부분의 상품들이 수입품이므로, 국제 유가 상승은 "Favorita"의 영향을 미칠 수 있습니다. 이러한 상황에서는 고객들이 지출을 줄이는 경향을 보일 수 있으며, "Favorita"의 매출과 이익에 부정적인 영향을 미칠 수 있습니다.
- 경쟁 업체 증가: 에콰도르의 소매 시장에서는 경쟁 업체들의 증가로 인해 "Favorita"의 시장 점유율이 감소할 가능성이 있습니다. 이러한 상황에서는 "Favorita"가 더욱 높은 고객 경험과 추가적인 가치를 제공할 필요가 있습니다.

```python
# 결측치 보간
oil_inter = pd.Series(oil['dcoilwtico']).interpolate()
oil['dcoilwtico_inter'] = oil_inter

oil['date'] = pd.to_datetime(oil['date'])

import pandas as pd

# 2013년 1월 1일부터 2017년 8월 31일까지 모든 날짜를 포함하는 DatetimeIndex를 생성합니다.
date_range = pd.date_range(start='2013-01-01', end='2017-08-31', freq='D')

# DatetimeIndex를 인덱스로 갖는 데이터프레임을 생성합니다.
new_oil = pd.DataFrame(index=date_range).reset_index()
new_oil = new_oil.rename(columns={'index':'date'})

# 데이터프레임을 출력합니다.
print(new_oil)

new_oil = new_oil.merge(oil, how='left', on='date')
new_oil

#주말 데이터 금요일 유가로 채우기
new_oil['dcoilwtico_inter'] = new_oil['dcoilwtico_inter'].fillna(method='ffill')

# 2013-01-01 에 데이터 채우기
# 다음 날의 데이터로 NaN 값 채우기
new_oil['dcoilwtico_inter'].fillna(method='bfill', inplace=True)
```

```python
train_grouped = train.groupby(['date'])['sales'].sum().reset_index()
train_grouped
```

```python
train_oil = pd.merge(train_grouped, new_oil, how='left' )

# 'dcoilwtico' 열 삭제하기
train_oil.drop('dcoilwtico', axis=1, inplace=True)

train_oil
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12, 6))

# 왼쪽 축 (sales)에 대한 데이터 플롯
ax1.plot(train_oil['date'], train_oil['sales'], color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Sales', color='blue')

ax2 = ax1.twinx()  # 오른쪽 축 생성

# 오른쪽 축 (dcoilwtico_inter)에 대한 데이터 플롯
ax2.plot(train_oil['date'], train_oil['dcoilwtico_inter'], color='red')
ax2.set_ylabel('Oil Price', color='red')

plt.show()
```

```python
import pandas as pd
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_app")

# "city", "state", "country" 열의 정보를 이용해 위도와 경도를 찾아 새로운 열로 추가합니다.
stores["location"] = stores.apply(lambda row: geolocator.geocode(row["city"] + ", " + row["state"] + ", " + row["country"]), axis=1)
stores["latitude"] = stores["location"].apply(lambda loc: loc.latitude if loc else None)
stores["longitude"] = stores["location"].apply(lambda loc: loc.longitude if loc else None)

# 결과를 출력합니다.
stores.head(2)
```

```python
import folium

# 지도 초기화
m = folium.Map(location=[-0.180653, -78.467834], zoom_start=7)

# 클러스터별 폴리곤 생성
for cluster in stores['cluster'].unique():
    cluster_data = stores[stores['cluster'] == cluster]
    polygon_coords = list(zip(cluster_data['latitude'], cluster_data['longitude']))
    polygon = folium.Polygon(locations=polygon_coords, fill_color='blue', fill_opacity=0.3, stroke=False)
    m.add_child(polygon)

# 매장 위치 마커 생성
for index, row in stores.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']]).add_to(m)

# 지도 출력
m
```

## tensorflow

```python
input_shape = X_train.shape[1]
input_shape
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=128, activation='relu', 
                          input_shape=[input_shape]),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dropout(0.05),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dropout(0.05),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
  ])

model.summary()
```

```python
(input_shape * 64)  + 64
```

```python
tf.keras.losses.BinaryCrossentropy()
```

```python
# 모델 컴파일
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, 
              loss="binary_crossentropy",
              metrics=["accuracy"]
              )
```

```python
class PrintDot(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')

# val_loss 기준으로 값이 나아지지 않으면 멈추게 합니다.
early_stop = tf.keras.callbacks.EarlyStopping(monitor='loss', 
                                              patience=100,)
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=100)

history = model.fit(X_train, y_train, 
                    epochs=50, validation_data=(X_valid, y_valid), 
                    batch_size=10,
                    callbacks=[early_stop], verbose=1)
history
```
